# BAR plot

```python
x=['python','c','c++','java']
y=[85,70,60,82]
c =['r','b','k','y']
plt.xlabel("Language")
plt.ylabel("poplularity")
plt.title("lang-pop") # graph se bahr naam dega 
plt.bar(x,y,width=0.5,color=c,edgecolor='k',linewidth=3,label='legend',alpha=0.5)  #alpha dims the colour 
plt.legend() #put that legend in graph which is connected to lable in plt.bar
plt.show()
```
